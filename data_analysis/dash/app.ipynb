```python
@capture("graph")
def tree_type(data_frame):
    fig = px.treemap(data_frame, path=['sector', 'type'],values='count',custom_data=["sector","document_id"])
    fig.data[0].textinfo = 'value'
    return fig
```

```python
Vizro._reset()
page_vendas = vm.Page(
    title="Vendas",
    layout=vm.Layout(grid=[[0,0,0],[1,1,2]]),
    components=[
        
        vm.Graph(id="growth", figure=px.bar(monthly_growth_empresa, x="date_time", y="value",color="variable",barmode="group",custom_data=["sector","document_id","variable"])),
        vm.Graph(id="kind", figure=tree_type(kind)),
        vm.Graph(id="geo", figure=px.pie(state,names="state",values="count",custom_data=["sector","document_id"])),
],
    controls=[
        
        vm.Filter(column="sector", selector=vm.Dropdown(value=["ALL"])),
        vm.Filter(column="document_id", selector=vm.Dropdown(value=["1898616423937443"])),
        vm.Filter(column="variable", selector=vm.Dropdown()),
        vm.Filter(column="date_time"),
    ],
)

page_liquidez = vm.Page(
    title="Liquidez",
    layout=vm.Layout(grid=[[0, 0],[1,2]]),
    components=[
        vm.Graph(id="cumsum", figure=px.line(cum_day, x="date_time", y="value",color="id_var",custom_data=["sector","document_id","variable"]),actions=[vm.Action(function=filter_interaction(targets=["in","out"]))]),
        vm.Graph(id="in", figure=px.line(inflow_day, x="date_time", y="in",color='document_id',custom_data=["sector","document_id"])),
        vm.Graph(id="out", figure=px.line(outflow_day, x="date_time", y="out",color='document_id',custom_data=["sector","document_id"])),
],
    controls=[
        
        vm.Filter(column="sector", selector=vm.Dropdown(value=["Bakeries"])),
        vm.Filter(column="document_id", selector=vm.Dropdown(value=["1898616423937443"])),
        vm.Filter(column="variable", selector=vm.Dropdown()),
        vm.Filter(column="date_time"),
    ],
)

dashboard = vm.Dashboard(pages=[page_liquidez,page_vendas],theme= "vizro_light")

Vizro().build(dashboard).run(jupyter_mode="external",port=6969)
```
